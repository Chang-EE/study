```
image = np.zeros((500,500,3),dtype="uint8")
image[:] = colors['light_gray']
separation = 40

for key in colors:
  cv2.line(image, (0,separation), (500, separation), colors[key], 10)
  separation += 40

img_RGB = image[:, :, ::-1]
plt.imshow(img_RGB)
plt.show()
```

```
def my_flip(image, opt):
  img = image

  if opt == 0:
    for i in range(img.shape[0]):
      img[i,:] , img[img.shape[0]-i-1,:] = img[img.shape[0]-i-1,:] , img[i,:]

  elif opt == 1:
    for i in range(img.shape[1]):
      img[:,i] , img[:,img.shape[1]-i-1] = img[:,img.shape[1]-i-1] , img[:,i]

  elif opt == -1:
    for i in range(img.shape[0]):
      for j in range(img.shape[1]):
        img[i,j] , img[img.shape[0]-i-1,img.shape[1]-j-1] = img[img.shape[0]-i-1,img.shape[1]-i-1]

  return img


def my_repeat(image, y_rep, x_rep):
  img = image
  exp_img = np.zeros((img.shape[0]*y_rep, img.shape[1]*x_rep, img.shape[2]))
  for i in range(img.shape[0]):
    for j in range(img.shape[1]):
      for k in range(img.shape[2]):
        exp_img[i*y_rep:(i+1)*y_rep, j*x_rep:(j+1)*x_rep, k] = img[i,j,k]
  return


def my_transpose(image):
  img = image

image = cv2.imread("flip_test.jpg", cv2.IMREAD_COLOR)
if image is None: raise Exception("    ") #  

x_axis = my_flip(image, 0)                 # x   
y_axis = my_flip(image, 1)                 # y   
xy_axis = my_flip(image, -1)
#rep_image   = cv2.repeat(image, 1, 2)       #  
#trans_image = cv2.transpose(image)          #  


```
