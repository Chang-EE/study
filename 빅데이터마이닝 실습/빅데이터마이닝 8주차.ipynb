```python
def AND(x1,x2):
  w1, w2, theta = 0.5, 0.5, 0.7
  tmp = w1*x1 + w2*x2
  if tmp <= theta:
    return 0
  elif tmp > theta:
    return 1

print(f" 0 and 0: {AND(0,0)}\n 0 and 1: {AND(0,1)}\n 1 and 0: {AND(1,0)}\n 1 and 1: {AND(1,1)}\n")
```

```python
def AND(x1,x2):
  x = np.array([x1,x2])
  w = np.array([0.5,0.5])
  b = -0.7
  tmp = np.sum(w*x) + b

  if tmp <= 0:
    return 0
  else:
    return 1

print(f" 0 and 0: {AND(0,0)}\n 0 and 1: {AND(0,1)}\n 1 and 0: {AND(1,0)}\n 1 and 1: {AND(1,1)}\n")
```

```python
def NAND(x1,x2):
  x = np.array([x1,x2])
  w = np.array([-1,-1])
  b = 1.9
  tmp = np.sum(w*x) + b

  if tmp <= 0:
    return 0
  else:
    return 1

print(f" 0 nand 0: {NAND(0,0)}\n 0 nand 1: {NAND(0,1)}\n 1 nand 0: {NAND(1,0)}\n 1 nand 1: {AND(1,1)}\n")
```

```python
def OR(x1,x2):
  x = np.array([x1,x2])
  w = np.array([1,1])
  b = 0
  tmp = np.sum(w*x) + b

  if tmp <= 0:
    return 0
  else:
    return 1

print(f" 0 or 0: {OR(0,0)}\n 0 or 1: {OR(0,1)}\n 1 or 0: {OR(1,0)}\n 1 or 1: {OR(1,1)}\n")
```

```python
def XOR(x1,x2):
  return AND(NAND(x1,x2),OR(x1,x2))

print(f" 0 xor 0: {XOR(0,0)}\n 0 xor 1: {XOR(0,1)}\n 1 xor 0: {XOR(1,0)}\n 1 xor 1: {XOR(1,1)}\n")
```
